```python
# Import all the image files

import os,glob

categories = ["Costa coffee","Costco", "Hobby lobby", "Starbucks", "Target"]
path = f'Data\images\{categories[0]}'
files = os.listdir(path)
count = 0

for f in files:
    count += 1
#     print(f)
    
print("Total number of files in {} are {}".format(categories[0], count))
    
```

```python
count = 0

for cat in categories:
    path = f'Data\images\{cat}'
    files = os.listdir(path)
    
    for f in files:
        if(os.path.splitext(f)[1] == ".jpg"):
            count +=1
            print(f)
        else:
            next
            
print("Total number of images are {}".format(count))
```
