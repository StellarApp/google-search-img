```python
# Google-search API
# Extract the image url from the searched data

def get_data(keyword, pg_num):
    params = {
        "q": keyword,
        "tbm": "isch",
        "api_key": SERP_API_KEY,
        "output": "json"
    }

    params["ijn"] = pg_num
    search = GoogleSearch(params)
    results = search.get_dict()
    
    urls = []
    
    for item in results['images_results']:
        try:
            urls.append(item['original'])
        except: 
            pass
    return urls
```

```python
get_data("starbucks storefront", 2)[:5]
```

```python
img_urls = []

for i in range(2):
    img_urls += get_data("starbucks storefront", i)
```

```python
# Collect the images for each keyword

keywords = ['Carters', 'Costa']

for keyword in keywords:
    img_urls = []
    dir_path = f'Data/images/{keyword}/{keyword}_'
    
    img_urls += get_data(f'{keyword} storefront', 1)

    for i,url in enumerate(img_urls[:30]):
        try: 
            wget.download(url, dir_path + str(i) + ".jpg")
        except:
            pass
```
